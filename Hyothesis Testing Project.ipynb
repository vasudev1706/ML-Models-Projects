LABTAT DATASET

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df=pd.read_csv("/content/LabTAT.csv")
```

```python
df
```

```python
lab1= df["Laboratory 1"]
lab2= df["Laboratory 2"]
lab3= df["Laboratory 3"]
lab4 = df["Laboratory 4"]
```

```python
# Performing one-way ANOVA
f_statistic, p_value = stats.f_oneway(lab1, lab2, lab3, lab4)
```

```python
# Setting the significance level
alpha = 0.05
```

```python
#results
```

```python
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

```python
# Checking the p-value against the significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in average TAT.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in average TAT.")
```

CUTLETS DATASET

```python
df1=pd.read_csv("/content/Cutlets.csv")
```

```python
df1
```

```python
unit_a= df1["Unit A"]
unit_b= df1["Unit B"]
```

```python
# Assumption checks
_, p_value_a = stats.shapiro(unit_a)
_, p_value_b = stats.shapiro(unit_b)

_, p_value_var = stats.levene(unit_a, unit_b)
```

```python
# Two-sample t-test
t_statistic, p_value_ttest = stats.ttest_ind(unit_a, unit_b)
```

```python
# Display results
print(f"Shapiro-Wilk test for normality (Unit A): p-value = {p_value_a:.4f}")
print(f"Shapiro-Wilk test for normality (Unit B): p-value = {p_value_b:.4f}")
print(f"Levene's test for homogeneity of variance: p-value = {p_value_var:.4f}")

print("\nTwo-sample t-test results:")
print(f"t-statistic = {t_statistic:.4f}")
print(f"p-value = {p_value_ttest:.4f}")
```

```python
# Interpret the results
alpha = 0.05
if p_value_a > alpha and p_value_b > alpha and p_value_var > alpha:
    print("\nAssumptions met. There is no significant difference.")
    print("Fail to reject the null hypothesis.")
else:
    print("\nAssumptions not met or p-value < 0.05. There is a significant difference.")
    print("Reject the null hypothesis.")
```

CUSTOMER ORDERFORM DATASET


```python
df3= pd.read_csv("/content/Costomer+OrderForm.csv")
```

```python
df3
```

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(df3["Phillippines"], [df3['Indonesia'], df3['Malta'], df3['India']], margins=True)
```

```python
contingency_table
```

```python
# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
```

```python
# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies table:")
print(expected)
```

```python
# Compare p-value to significance level (alpha)
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis. There is a significant association between center and defective/error-free status.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between center and defective/error-free status.")
```
