```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("wine.csv")
df=df.drop("Type",axis=1)
```

```python
#importing PCA
```

```python
from sklearn.decomposition import PCA
```

```python
pc=PCA(n_components=3)
```

```python
pc.fit_transform(df)
```

```python
pca=pd.DataFrame(pc.fit_transform(df),columns=["pc1","pc2","pc3"])
```

```python
pca
```

```python
pca1=pca.copy()
```

```python
#variance for pca's
var=pc.explained_variance_ratio_
var
```

```python
var1=np.cumsum(var*100)
```

KMEANS

```python
#finding out number clusters
```

```python
pca.iloc[:,:3]
```

```python
#importing kmeans
from sklearn.cluster import KMeans
```

```python
#inertia calculates the within cluster sum of squares for clusters from 1 to 11
```

```python
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(pca.iloc[:,:3])
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
#visualization for better understanding
sns.lineplot(x=range(1,11),y=wcss)
```

```python
#as we can see the line is dropping at a slow rate after 3, so we can take 3 as cluster
```

```python
kmeans=KMeans(n_clusters=3)
predict=kmeans.fit_predict(pca.iloc[:,:3])
```

```python
predict
```

```python
pca1["cluster1"]=kmeans.labels_
```

```python
pca1
```

```python
kmeans.cluster_centers_
```

```python
#visualization
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(pca1['pc1'], pca1['pc2'], pca1['pc3'], c=pca1['cluster1'], cmap='viridis', s=80)
ax.set_title('3D Scatter Plot of Principal Components with Clusters')
ax.set_xlabel('pc1')
ax.set_ylabel('pc2')
ax.set_zlabel('pc3')
plt.legend(*scatter.legend_elements(), title='Clusters')
plt.show()
```
